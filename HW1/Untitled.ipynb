```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./Customer_Churn.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
corr_map = df.corr()
sns.set(font_scale=1.1)
sns.heatmap(corr_map).set(title='Correlation of all features')
```

```python
sns.histplot(df['INCOME'],kde='True').set(title='Distribution of Income')
plt.axvline(x=df['INCOME'].mean(),
            color='red', label='mean',ls='--')
plt.axvline(x=df['INCOME'].median(),
            color='green', label='median')
plt.legend()
```
